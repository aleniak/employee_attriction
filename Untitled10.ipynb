```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Настройка отображения графиков
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (10, 6)
sns.set_palette("husl")

# Загрузка данных
df = pd.read_csv('Attrition Data.csv')

```

```python

print("=" * 50)
print("ОСНОВНАЯ ИНФОРМАЦИЯ О ДАННЫХ")
print("=" * 50)
print(f"Размер датасета: {df.shape}")
print(f"Количество строк: {df.shape[0]}")
print(f"Количество столбцов: {df.shape[1]}")

print("\n" + "=" * 50)
print("ИНФОРМАЦИЯ О СТОЛБЦАХ")
print("=" * 50)
print(df.info())
```

```python
print("\n" + "=" * 50)
print("ПЕРВЫЕ 5 СТРОК")
print("=" * 50)
print(df.head())

print("\n" + "=" * 50)
print("ОСНОВНЫЕ СТАТИСТИКИ")
print("=" * 50)
print(df.describe())
```

```python
print("\n" + "=" * 50)
print("ПРОВЕРКА НА ПРОПУЩЕННЫЕ ЗНАЧЕНИЯ")
print("=" * 50)
print(df.isnull().sum())

print("\n" + "=" * 50)
print("ПРОВЕРКА НА ДУБЛИКАТЫ")
print("=" * 50)
print(f"Количество дубликатов: {df.duplicated().sum()}")
```

```python
# Анализ целевой переменной Attrition
print("\n" + "=" * 50)
print("АНАЛИЗ ЦЕЛЕВОЙ ПЕРЕМЕННОЙ ATTRITION")
print("=" * 50)
attrition_counts = df['Attrition'].value_counts()
attrition_percent = df['Attrition'].value_counts(normalize=True) * 100

print("Распределение Attrition:")
for i, (count, percent) in enumerate(zip(attrition_counts, attrition_percent)):
    print(f"  {attrition_counts.index[i]}: {count} ({percent:.2f}%)")

# Визуализация целевой переменной
plt.figure(figsize=(10, 5))
```

```python
plt.subplot(1, 2, 1)
plt.pie(attrition_counts.values, labels=attrition_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Распределение Attrition')

plt.subplot(1, 2, 2)
sns.countplot(data=df, x='Attrition')
plt.title('Количество случаев Attrition')
plt.tight_layout()
plt.show()
```

```python
# Анализ категориальных переменных
print("\n" + "=" * 50)
print("АНАЛИЗ КАТЕГОРИАЛЬНЫХ ПЕРЕМЕННЫХ")
print("=" * 50)

categorical_columns = ['Department', 'EducationField', 'MaritalStatus']
for col in categorical_columns:
    print(f"\n--- {col} ---")
    print(df[col].value_counts())

    # Визуализация
    plt.figure(figsize=(10, 4))

    plt.subplot(1, 2, 1)
    sns.countplot(data=df, x=col)
    plt.title(f'Распределение {col}')
    plt.xticks(rotation=45)

    plt.subplot(1, 2, 2)
    sns.countplot(data=df, x=col, hue='Attrition')
    plt.title(f'{col} по Attrition')
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()
```

```python
print("\n" + "=" * 50)
print("АНАЛИЗ ЧИСЛОВЫХ ПЕРЕМЕННЫХ")
print("=" * 50)

numeric_columns = ['Age', 'DistanceFromHome', 'Education', 'EnvironmentSatisfaction',
                  'JobSatisfaction', 'MonthlyIncome', 'NumCompaniesWorked',
                  'WorkLifeBalance', 'YearsAtCompany']

# Гистограммы для числовых переменных
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.ravel()

for i, col in enumerate(numeric_columns):
    df[col].hist(bins=30, ax=axes[i])
    axes[i].set_title(f'Распределение {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Частота')

plt.tight_layout()
plt.show()

# Боксплоты для числовых переменных по Attrition
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.ravel()

for i, col in enumerate(numeric_columns):
    sns.boxplot(data=df, x='Attrition', y=col, ax=axes[i])
    axes[i].set_title(f'{col} по Attrition')

plt.tight_layout()
plt.show()
```

```python
print("\n" + "=" * 50)
print("КОРРЕЛЯЦИОННЫЙ АНАЛИЗ")
print("=" * 50)

# Создаем числовую версию датасета для корреляции
df_numeric = df.copy()
df_numeric['Attrition'] = df_numeric['Attrition'].map({'Yes': 1, 'No': 0})

# Кодируем категориальные переменные
for col in categorical_columns:
    df_numeric[col] = pd.factorize(df_numeric[col])[0]

plt.figure(figsize=(12, 10))
correlation_matrix = df_numeric.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            fmt='.2f', linewidths=0.5)
plt.title('Матрица корреляций')
plt.tight_layout()
plt.show()

# Анализ наиболее коррелированных переменных с Attrition
print("\nКорреляция с Attrition:")
attrition_corr = correlation_matrix['Attrition'].sort_values(ascending=False)
print(attrition_corr)
```

```python

print("\n" + "=" * 50)
print("ДЕТАЛЬНЫЙ АНАЛИЗ КЛЮЧЕВЫХ ПЕРЕМЕННЫХ")
print("=" * 50)

# Анализ MonthlyIncome по Attrition
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome')
plt.title('Monthly Income по Attrition')

plt.subplot(1, 2, 2)
sns.histplot(data=df, x='MonthlyIncome', hue='Attrition', kde=True)
plt.title('Распределение Monthly Income по Attrition')

plt.tight_layout()
plt.show()
```

```python
income_yes = df[df['Attrition'] == 'Yes']['MonthlyIncome']
income_no = df[df['Attrition'] == 'No']['MonthlyIncome']
t_stat, p_value = stats.ttest_ind(income_yes, income_no)
print(f"T-тест для MonthlyIncome: t={t_stat:.3f}, p={p_value:.5f}")

# Анализ Age по Attrition
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='Attrition', y='Age')
plt.title('Age по Attrition')

plt.subplot(1, 2, 2)
sns.histplot(data=df, x='Age', hue='Attrition', kde=True)
plt.title('Распределение Age по Attrition')

plt.tight_layout()
plt.show()

# Анализ YearsAtCompany по Attrition
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='Attrition', y='YearsAtCompany')
plt.title('YearsAtCompany по Attrition')

plt.subplot(1, 2, 2)
sns.histplot(data=df, x='YearsAtCompany', hue='Attrition', kde=True)
plt.title('Распределение YearsAtCompany по Attrition')

plt.tight_layout()
plt.show()

# Анализ удовлетворенности
satisfaction_columns = ['EnvironmentSatisfaction', 'JobSatisfaction', 'WorkLifeBalance']

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, col in enumerate(satisfaction_columns):
    pd.crosstab(df[col], df['Attrition'], normalize='index').plot(
        kind='bar', ax=axes[i], title=f'{col} vs Attrition'
    )
    axes[i].legend(title='Attrition')
    axes[i].set_xlabel(col)

plt.tight_layout()
plt.show()
```

```python
# Анализ по Department и EducationField
print("\n" + "=" * 50)
print("АНАЛИЗ ПО ДЕПАРТАМЕНТАМ И ОБРАЗОВАНИЮ")
print("=" * 50)

# Attrition rate по департаментам
dept_attrition = pd.crosstab(df['Department'], df['Attrition'], normalize='index') * 100
print("Attrition rate по департаментам (%):")
print(dept_attrition['Yes'].sort_values(ascending=False))

# Attrition rate по образованию
edu_attrition = pd.crosstab(df['EducationField'], df['Attrition'], normalize='index') * 100
print("\nAttrition rate по EducationField (%):")
print(edu_attrition['Yes'].sort_values(ascending=False))

# Визуализация
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

dept_attrition['Yes'].sort_values().plot(kind='barh', ax=axes[0])
axes[0].set_title('Attrition Rate по Department')
axes[0].set_xlabel('Attrition Rate (%)')

edu_attrition['Yes'].sort_values().plot(kind='barh', ax=axes[1])
axes[1].set_title('Attrition Rate по EducationField')
axes[1].set_xlabel('Attrition Rate (%)')

plt.tight_layout()
plt.show()

# Анализ выбросов
print("\n" + "=" * 50)
print("АНАЛИЗ ВЫБРОСОВ")
print("=" * 50)

def detect_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = column[(column < lower_bound) | (column > upper_bound)]
    return len(outliers)

print("Количество выбросов по переменным:")
for col in numeric_columns:
    outliers_count = detect_outliers(df[col])
    print(f"  {col}: {outliers_count} выбросов")
```

```python
print("\n" + "=" * 50)
print("ФИНАЛЬНЫЙ ОТЧЕТ EDA")
print("=" * 50)
print(f"• Всего наблюдений: {df.shape[0]}")
print(f"• Attrition rate: {attrition_percent['Yes']:.2f}%")
print(f"• Самые важные предикторы Attrition:")
for i, (var, corr) in enumerate(attrition_corr[1:4].items()):
    print(f"    {i+1}. {var}: {corr:.3f}")
print(f"• Департамент с самым высоким Attrition: {dept_attrition['Yes'].idxmax()} ({dept_attrition['Yes'].max():.1f}%)")
print(f"• Поле образования с самым высоким Attrition: {edu_attrition['Yes'].idxmax()} ({edu_attrition['Yes'].max():.1f}%)")
```
